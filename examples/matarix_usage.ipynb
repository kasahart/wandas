```python
import numpy as np
from wandas.core.matrix_frame import MatrixFrame
```

# MatrixFrame

```python
# サンプルデータの作成
num_channels = 2
num_samples = 16000
sampling_rate = 16000
data = np.random.randn(num_channels, num_samples)  # ランダムなデータ

metadata = [
    {'sensor': 'microphone', 'position': 'left'},
    {'sensor': 'microphone', 'position': 'right'},
]

# MatrixFrame の作成
matrix_frame = MatrixFrame(
    data=data,
    sampling_rate=sampling_rate,
    channel_labels=['Ch1', 'Ch2'],
    channel_metadata=metadata,
    channel_units=['Pa', 'V'],
    label='Test Signal'
)
```

```python
matrix_frame.plot(title="Original Signal", overlay=False)
```

```python
matrix_frame["Ch1"].describe()
```

```python
matrix_frame.coherence().plot()
```

```python
matrix_frame.coherence().plot_matrix()
```

```python
matrix_frame.csd().plot()
```

```python
matrix_frame.csd().plot_matrix()
```
