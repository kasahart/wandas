```python
import pandas as pd

from wandas.core import ChannelFrame
```

# wavファイル読み込み

```python
cf = ChannelFrame.read_wav("data/summer_streets1.wav")
cf.describe()
```

軸の設定

```python
cf.describe(
    axis_config={
        "time_plot": {"xlim": (0, 15), "ylim": (-30000, 30000)},
        "freq_plot": {"xlim": (60, 120), "ylim": (0, 16000)},
    },
    cbar_config={"vmin": 10, "vmax": 70},
)
```
