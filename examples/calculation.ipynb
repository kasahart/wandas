```python
import wandas as wd
```

```python
# Generate a sample
signal = wd.generate_sin(freqs=[1000, 5000], duration=2)
signal.plot()
signal.describe()
```

```python
# 定数との計算
(signal[0] * 20).describe()
```

```python
# chnnel間の計算
(signal[0] + signal[0] * -1).describe()
```

```python
# ndarrayとの計算
(signal[0] + signal[0].data * -1).describe()
```

```python
(signal[0].add(signal[1])).describe()
```

```python
(signal[0].add(signal[1], snr=10)).describe()
```

```python
(signal[0].add(signal[1], snr=-10)).describe()
```
